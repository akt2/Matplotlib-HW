```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
cd=pd.read_csv('data/city_data.csv')
rd=pd.read_csv('data/ride_data.csv')
# Combine the data into a single dataset
pb=pd.merge(rd,cd,how='left',on='city')
# Display the data table for preview
pb.head()
pb.set_index('city').head()
```

```python
pbg=pb.groupby(['city','type'], as_index=False)
# xpb=pbg['ride_id'].count()
# ypb=pbg['fare'].mean()
# apb=pbg['driver_count']

pyb=pbg.agg({'ride_id': 'nunique','fare':'mean','driver_count':'mean'})
pyb.head()
pyb=pyb.replace(['Urban','Suburban','Rural'],[1,2,3])
```

```python
x = pyb['ride_id']
y=pyb['fare']
s=pyb['driver_count']
c=pyb['type']

plt.scatter(x,y,s=s,c=c,alpha=0.5,label=('Urban','Suburban','Rural'))
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Pyber ride sharing data for 2016')
plt.savefig("../Images/PyberBubble.png")
```

## Total Fares by City Type

```python
# Calculate Type Percents
ptg=pb.groupby(['type'], as_index=False)

pyb=ptg.agg({'ride_id': 'nunique','fare':'sum','driver_count':'mean'})
colors=['gold','lightskyblue','lightcoral']
explode=(0,0,0.1)
pyb.head()
plt.pie(pyb['fare'],explode=explode,labels=pyb['type'],colors=colors,autopct='%1.2f%%',shadow=True,startangle=120)
plt.axis('equal')
plt.title('% Total Fares by City Type')
plt.show()
plt.savefig("../Images/PyberPieFares.png")
```
